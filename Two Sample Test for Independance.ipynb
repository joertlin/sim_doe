```python
%matplotlib inline
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
def reject_null_hyp(alpha,p_val):
    if p_val<alpha:
        return True
    else:
        return False
```

```python
def print_ttest_descriptives(a,b,test):
    print('Statistics for a and b')
    print('-----a-----')
    print('average = %0.2f'%np.average(a))
    print('standard deviation = %0.2f'%np.std(a))
    print('-----b-----')
    print('average = %0.2f'%np.average(b))
    print('standard deviation = %0.2f'%np.std(b))
    print('---------------')
    print('Test for independant sample:')
    print('t-Statistic = %0.3f'%test[0])
    print('p-Value = %0.3f'%test[1])
```

Let's take a look at statistical tests for independance.  
---
Numpy is used for all of the functions below.

First, set up a couple of sample functions to obtain random normal values.

Second, perform a t-test for independant samples

**Null Hypothesis**

H0: Sample a and sample b have the same mean value.

H1: Sample a and sample b have different means.

```python
size = 10000
offset = 0.
a = np.random.randn(size) + offset
b = np.random.randn(size) + offset

test = stats.ttest_ind(a,b)

plt.scatter(a,b)
plt.xlabel('a')
plt.ylabel('b')
plt.title('Scatter plot of a vs. b')
plt.axis('equal')
plt.show()

print_ttest_descriptives(a,b,test)
```

Results
---
p-Value is greater than alpha (assume alpha = 0.05). We have no reason to reject the null hypothesis.  The random normal sample sets generated from the same function appear to have the statistically equal mean values. This is expected....

(Note that there is a small change, approximately equal to alpha, that the above statement will be incorrectly stated.)

What would happen if we use the same function but ran the test several hundred times?

```python
t_vals = []
p_vals = []
n_test = 1000
size = 60
offset = 100.
spread = .1*offset
for i in range(n_test):
    a = np.random.randn(size) * spread + offset
    b = np.random.randn(size) * spread + offset
    
    result_ttest = stats.ttest_ind(a,b)
    
    t_vals.append(result_ttest[0])
    p_vals.append(result_ttest[1])
```

```python
plt.hist(p_vals)
plt.xlabel('p - Value')
plt.title('Histogram of %i tests for independance\n between two random samples generated\n with the same functuon.'%n_test)
plt.show()
```

```python
alpha = 0.05
n_true = [reject_null_hyp(alpha,p_val) for p_val in p_vals].count(True)
n_false = len(p_vals) - n_true
plt.bar([0,1],[n_false,n_true],align = 'center', width = .25)
plt.xticks([0,1],['False','True'])

percent = float(n_true)/float(len(p_vals))*100.
print('We reject the null hypothesis %i of %i tests(%0.1f%% of the time).'%(n_true,len(p_vals),percent))
```

**Test Cases**

In this setup we have designed a function to generate a random normal distribution for a and b. The equation is identical, so we would expect a and b to be from the sampe population.

At a confidence level of 5% (alpha = 0.05) we would expect to reject the null hypothesis 5% of the time by random chance.  In the example run, we have *n_test* tests run with a sample size of *size*.
